```python
# Block: derive nx/ny/nk/step from native jampy grid logic, then inject into JAX JAM
import numpy as np

nrad_use = int(globals().get('nrad', 20))
nang_use = int(globals().get('nang', 10))


def derive_base_grid_params(x, y, sigma_lum, qobs_lum, sigma_psf, pix_size,
                            nrad=20, nang=10, interp=True, step_input=0.0):
    # Match native jampy interpolation/PSF grid-size logic.
    x = np.asarray(x)
    y = np.asarray(y)
    sigma_lum = np.asarray(sigma_lum)
    qobs_lum = np.asarray(qobs_lum)
    sigma_psf = np.asarray(sigma_psf)

    w = sigma_lum < np.max(np.abs(x))
    qmed = np.median(qobs_lum) if w.sum() < 3 else np.median(qobs_lum[w])
    rell = np.hypot(x, y / qmed)

    psf_convolution = (np.max(sigma_psf) > 0) and (pix_size > 0)

    if (not interp) or ((nrad * nang > x.size) and (not psf_convolution)):
        step = np.min(rell) if step_input == 0 else step_input
        mx = 0.0
    else:
        if psf_convolution:
            step = np.min(sigma_psf) / 4 if step_input == 0 else step_input
            mx = 3 * np.max(sigma_psf) + pix_size / np.sqrt(2)
        else:
            step = np.min(rell) if step_input == 0 else step_input
            mx = 0.0

    rmax = np.max(rell) + mx
    nx = int(np.ceil(rmax / step))
    ny = int(np.ceil(rmax * qmed / step))
    nk = int(np.ceil(mx / step))
    return nx, ny, nk, float(step), float(qmed), float(rmax), float(mx)


nx_base, ny_base, nk_base, step_base, qmed_base, rmax_base, mx_base = derive_base_grid_params(
    x=xbin,
    y=ybin,
    sigma_lum=sigma_lum,
    qobs_lum=qobs_lum,
    sigma_psf=sigmapsf,
    pix_size=pixsize,
    nrad=nrad_use,
    nang=nang_use,
    interp=True,
    step_input=0.0,
)

print(f'[derived-base-grid] nx={nx_base}, ny={ny_base}, nk={nk_base}, step={step_base:.6g}')
print(f'[derived-base-grid] qmed={qmed_base:.6g}, rmax={rmax_base:.6g}, mx={mx_base:.6g}')

# One explicit JAX JAM call using the derived grid settings
model_basegrid, chi2_basegrid, flux_basegrid, ml_basegrid, vel_basegrid, vel2_basegrid = jam_obj.get_kinematics(
    surf_lum=surf_lum,
    sigma_lum=sigma_lum,
    qobs_lum=qobs_lum,
    surf_pot=surf_pot,
    sigma_pot=sigma_pot,
    qobs_pot=qobs_pot,
    inc=inc,
    mbh=mbh,
    distance=distance,
    xbin=xbin,
    ybin=ybin,
    data=rms,
    sigmapsf=sigmapsf,
    normpsf=normpsf,
    beta=beta,
    pixsize=pixsize,
    logistic=False,
    moment='zz',
    goodbins=goodbins,
    align='cyl',
    ml=None,
    nx=nx_base,
    ny=ny_base,
    nk=nk_base,
    step=step_base,
    quiet=True,
)

print(f'[jax-with-derived-grid] chi2={float(np.asarray(chi2_basegrid)):.6f}, ml={float(np.asarray(ml_basegrid)):.6f}, mean(model)={float(np.asarray(model_basegrid).mean()):.6g}')

```

```python
# Block: base-vs-jax sigma_los comparison (native jampy vs jax jampy)
import os
import sys
import importlib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm

local_jampy_root = '/mnt/d/lensing/Jampy/jampy-8.1.4'
sys.path.insert(0, local_jampy_root)
print(f'using local jampy source: {local_jampy_root}')


import jampy
import jampy.axi.jam_axi_proj as jam_base_mod
jam_base_mod = importlib.reload(jam_base_mod)
from jampy.axi.jam_axi_proj import jam_axi_proj as jam_base

print(f'base jampy version: {jampy.__version__}')
print(f'grid injected to jax: nx={nx_base}, ny={ny_base}, nk={nk_base}, step={step_base:.6g}')
print('nrad = number of logarithmic radial points on polar interpolation grid')
print('nang = number of linear eccentric-anomaly points on polar interpolation grid')
print('[note] native base jampy does not take nx/ny/nk kwargs; jax call below uses injected nx/ny/nk.')

r_plane_cmp = jnp.sqrt(xbin**2 + (ybin/jnp.cos(jnp.radians(inc)))**2)
vr_cmp = 2000 * jnp.sqrt(r_plane_cmp) / (r_plane_cmp + a)
vel_obs_cmp = jnp.where(r_plane_cmp == 0, 0.0, vr_cmp * jnp.sin(jnp.radians(inc)) * xbin / r_plane_cmp)
errors_vrms = jnp.full_like(rms, jnp.median(rms) * 0.05)
errors_vel = jnp.full_like(vel_obs_cmp, 5.0)

common_kwargs_vrms = dict(
    surf_lum=surf_lum,
    sigma_lum=sigma_lum,
    qobs_lum=qobs_lum,
    surf_pot=surf_pot,
    sigma_pot=sigma_pot,
    qobs_pot=qobs_pot,
    inc=inc,
    mbh=mbh,
    distance=distance,
    xbin=xbin,
    ybin=ybin,
    data=rms,
    errors=errors_vrms,
    sigmapsf=sigmapsf,
    normpsf=normpsf,
    beta=beta,
    pixsize=pixsize,
    logistic=False,
    moment='zz',
    goodbins=goodbins,
    align='cyl',
    ml=None,
    nrad=20,
    nang=10,
    nlos=1500,
    step=step_base,
)

common_kwargs_vel = dict(common_kwargs_vrms, moment='z', data=vel_obs_cmp, errors=errors_vel)

# Base jampy expects numpy/scalar types.
scalar_keys = {'inc', 'mbh', 'distance', 'pixsize', 'step'}
pass_keys = {'logistic', 'moment', 'align', 'ml'}


def to_base_kwargs(kwargs):
    base_kwargs = {}
    for k, v in kwargs.items():
        if v is None:
            base_kwargs[k] = v
        elif k in scalar_keys:
            base_kwargs[k] = float(v)
        elif k in pass_keys:
            base_kwargs[k] = v
        elif k == 'goodbins':
            base_kwargs[k] = np.asarray(v, dtype=bool)
        else:
            base_kwargs[k] = np.asarray(v)
    return base_kwargs


base_kwargs_vrms = to_base_kwargs(common_kwargs_vrms)
base_kwargs_vel = to_base_kwargs(common_kwargs_vel)

jam_b_vrms = jam_base(**base_kwargs_vrms, plot=False, quiet=True)
vrms_base_cmp = np.asarray(jam_b_vrms.model)

jam_b_vel = jam_base(**base_kwargs_vel, plot=False, quiet=True)
vel_base_cmp = np.asarray(jam_b_vel.model)

# Compare sigma_los = sqrt(max(Vrms^2 - V_los^2, 0)).
jax_kwargs_vrms = dict(common_kwargs_vrms, nx=nx_base, ny=ny_base, nk=nk_base)
vrms_jax_cmp, chi2_jax_cmp, flux_jax_cmp, ml_jax_cmp, vel_jax_cmp, vel2_jax_cmp = jam_obj.get_kinematics(
    **jax_kwargs_vrms,
    quiet=True,
)
vrms_jax_cmp = np.asarray(vrms_jax_cmp)

jax_kwargs_vel = dict(common_kwargs_vel, nx=nx_base, ny=ny_base, nk=nk_base)
vel_jax_cmp, _, _, _, _, _ = jam_obj.get_kinematics(
    **jax_kwargs_vel,
    quiet=True,
)
vel_jax_cmp = np.asarray(vel_jax_cmp)

sigma_base_cmp = np.sqrt(np.clip(vrms_base_cmp**2 - vel_base_cmp**2, 0.0, None))
sigma_jax_cmp = np.sqrt(np.clip(vrms_jax_cmp**2 - vel_jax_cmp**2, 0.0, None))

eps = np.finfo(float).eps
pct_diff_cmp = 100.0 * (sigma_jax_cmp - sigma_base_cmp) / np.where(np.abs(sigma_base_cmp) > eps, np.abs(sigma_base_cmp), np.nan)
abs_diff_cmp = np.abs(sigma_jax_cmp - sigma_base_cmp)

vmin = np.nanmin([np.nanmin(sigma_base_cmp), np.nanmin(sigma_jax_cmp)])
vmax = np.nanmax([np.nanmax(sigma_base_cmp), np.nanmax(sigma_jax_cmp)])

x_arr = np.asarray(xbin)
y_arr = np.asarray(ybin)
x_unique = np.unique(x_arr)
y_unique = np.unique(y_arr)


def to_grid(values):
    values = np.asarray(values)
    grid = np.full((y_unique.size, x_unique.size), np.nan)
    ix = np.searchsorted(x_unique, x_arr)
    iy = np.searchsorted(y_unique, y_arr)
    grid[iy, ix] = values
    return grid

base_grid = to_grid(sigma_base_cmp)
jax_grid = to_grid(sigma_jax_cmp)
diff_grid = to_grid(pct_diff_cmp)
abs_grid = to_grid(abs_diff_cmp)

dmax = np.nanmax(np.abs(diff_grid))
admax = np.nanmax(abs_grid)


def add_top_cbar(fig, mappable, ax, title, label):
    try:
        cbar = fig.colorbar(mappable, ax=ax, location='top', fraction=0.03, pad=0.05, format='%.3g')
    except TypeError:
        cbar = fig.colorbar(mappable, ax=ax, orientation='horizontal', fraction=0.03, pad=0.12, format='%.3g')
    cbar.ax.xaxis.set_label_position('top')
    cbar.ax.xaxis.set_ticks_position('bottom')
    cbar.ax.xaxis.get_offset_text().set_visible(False)
    cbar.ax.tick_params(labelsize=8, pad=2)
    cbar.set_label(f'{title} | {label}', fontsize=9, labelpad=8)
    return cbar


# Compact layout: near-zero outer whitespace + square pcolormesh panels
xmin, xmax = float(np.min(x_unique)), float(np.max(x_unique))
ymin, ymax = float(np.min(y_unique)), float(np.max(y_unique))
span = max(xmax - xmin, ymax - ymin)
xc, yc = 0.5 * (xmin + xmax), 0.5 * (ymin + ymax)
xlim_sq = (xc - 0.5 * span, xc + 0.5 * span)
ylim_sq = (yc - 0.5 * span, yc + 0.5 * span)

fig, axes = plt.subplots(1, 3, figsize=(12, 4), constrained_layout=False)
fig.subplots_adjust(left=0.004, right=0.996, bottom=0.015, top=0.84, wspace=0.008)

pc0 = axes[0].pcolormesh(x_unique, y_unique, base_grid, shading='auto', cmap='jet', vmin=vmin, vmax=vmax)
axes[0].set_xlabel('x [arcsec]')
axes[0].set_ylabel('y [arcsec]')
add_top_cbar(fig, pc0, axes[0], 'Base jampy sigma_los model', 'sigma_los [km/s]')

pc1 = axes[1].pcolormesh(x_unique, y_unique, jax_grid, shading='auto', cmap='jet', vmin=vmin, vmax=vmax)
axes[1].set_xlabel('x [arcsec]')
axes[1].set_ylabel('')
axes[1].tick_params(labelleft=False)
add_top_cbar(fig, pc1, axes[1], 'JAX jampy sigma_los model', 'sigma_los [km/s]')

norm = TwoSlopeNorm(vmin=-dmax, vcenter=0.0, vmax=dmax) if dmax > 0 else None
pc2 = axes[2].pcolormesh(x_unique, y_unique, np.nan_to_num(diff_grid, nan=0.0), shading='auto', cmap='coolwarm', norm=norm)
axes[2].set_xlabel('x [arcsec]')
axes[2].set_ylabel('')
axes[2].tick_params(labelleft=False)
add_top_cbar(fig, pc2, axes[2], 'sigma_los difference (%) (JAX - Base)', '%')

for ax in axes:
    ax.set_xlim(*xlim_sq)
    ax.set_ylim(*ylim_sq)
    ax.set_aspect('equal', adjustable='box')
    ax.set_box_aspect(1)
    ax.tick_params(axis='both', which='both', labelsize=8, pad=1)
    ax.scatter(x_arr, y_arr, s=4, c='k', alpha=0.15, linewidths=0, zorder=3)

out_png = 'jax_vs_base_sigma_los_comparison.png'
fig.savefig(out_png, dpi=240, bbox_inches='tight', pad_inches=0.01)
print(f'saved figure: {out_png}')
plt.show()

finite_pct = pct_diff_cmp[np.isfinite(pct_diff_cmp)]
print('sigma_los = sqrt(max(Vrms^2 - V_los^2, 0))')
print(f'chi2 (jax/base, Vrms fit): {float(np.asarray(chi2_jax_cmp)):.6f} / {float(jam_b_vrms.chi2):.6f}')
print(f'ml   (jax/base, Vrms fit): {float(np.asarray(ml_jax_cmp)):.6f} / {float(jam_b_vrms.ml):.6f}')
print(f'|Δsigma_los| stats [km/s]: mean={abs_diff_cmp.mean():.6g}, median={np.median(abs_diff_cmp):.6g}, max={abs_diff_cmp.max():.6g}')
if finite_pct.size:
    print(f'Δ% stats: mean={finite_pct.mean():.6g}%, median={np.median(finite_pct):.6g}%, max|%|={np.max(np.abs(finite_pct)):.6g}%')
else:
    print('Δ% stats: all invalid (base model contains zeros everywhere).')



# Max-bias diagnostics
idx_max_abs = int(np.argmax(abs_diff_cmp))
print(
    f'max |Δsigma_los| at (x,y)=({float(np.asarray(xbin)[idx_max_abs]):.4g}, {float(np.asarray(ybin)[idx_max_abs]):.4g}) arcsec: '
    f'{float(abs_diff_cmp[idx_max_abs]):.6g} km/s '
    f'(base={float(sigma_base_cmp[idx_max_abs]):.6g}, jax={float(sigma_jax_cmp[idx_max_abs]):.6g})'
)

finite_mask = np.isfinite(pct_diff_cmp)
if np.any(finite_mask):
    idx_max_pct = int(np.nanargmax(np.abs(pct_diff_cmp)))
    print(
        f'max |Δ%| at (x,y)=({float(np.asarray(xbin)[idx_max_pct]):.4g}, {float(np.asarray(ybin)[idx_max_pct]):.4g}) arcsec: '
        f'{float(np.abs(pct_diff_cmp[idx_max_pct])):.6g}% '
        f'(signed={float(pct_diff_cmp[idx_max_pct]):.6g}%)'
    )
```

```python
# Block: base-vs-jax EPL sigma_los comparison with zero rotation
import importlib
import pickle as pkl
import sys
from pathlib import Path

import jax.numpy as jnp
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import TwoSlopeNorm

# Resolve local imports directly for this block so it can run independently.
local_jaxpy_root = str(Path('/mnt/d/lensing/jaxpy/jaxpy/jaxpy'))
sys.path.insert(0, local_jaxpy_root)

import param_util
from SLCOSMO import tool
from MGE_jax import MGE
from jam_axi_proj_jax import jam_axi_proj

plt.ioff()

sigmapsf = jnp.array([0.6, 1.2])
normpsf = jnp.array([0.7, 0.3])
pixsize = 0.8

local_jampy_root = '/mnt/d/lensing/Jampy/jampy-8.1.4'
sys.path.insert(0, local_jampy_root)

import jampy
import jampy.axi.jam_axi_proj as jam_base_mod
jam_base_mod = importlib.reload(jam_base_mod)
from jampy.axi.jam_axi_proj import jam_axi_proj as jam_base

file = 'lens_light_jackpot.pkl'
lens_light = pkl.load(open(file, 'rb'))[0]
surf_lum_epl = jnp.asarray(lens_light['amp'])
sigma_lum_epl = jnp.asarray(lens_light['sigma'])
_, qobs_lum_epl = param_util.ellipticity2phi_q(lens_light['e1'], lens_light['e2'])
qobs_lum_epl = jnp.asarray(qobs_lum_epl)
beta_epl = jnp.zeros_like(surf_lum_epl)

x_lin_epl = np.linspace(-2.0, 2.0, 60)
y_lin_epl = np.linspace(-2.0, 2.0, 60)
x_mesh_epl, y_mesh_epl = np.meshgrid(x_lin_epl, y_lin_epl, indexing='xy')
xbin_epl = x_mesh_epl.ravel()
ybin_epl = y_mesh_epl.ravel()
goodbins_epl = np.ones_like(xbin_epl, dtype=bool)

cosmology_epl = {'Omegam': 0.3, 'Omegak': 0.0, 'w0': -1.0, 'wa': 0.0, 'h0': 70.0}
kwargs_epl_sigma = {'thetaE': 1.0, 'gamma': 2.0, 'zl': 0.2, 'zs': 0.4, 'cosmology': cosmology_epl}
dl_epl, ds_epl, dls_epl = tool.dldsdls(0.2, 0.4, cosmology_epl, n=20)
distance_epl = float(np.asarray(dl_epl).ravel()[0])
mge_epl_sigma = MGE(tool.EPL_msunmpc, 'thetaE', n_gauss=10, n_terms=28, sigma_start_mult=1/100, sigma_end_mult=50)
mass_amp_epl, mass_sig_epl = mge_epl_sigma.decompose(**kwargs_epl_sigma)

# Use a fixed projected mass flattening q=0.8 for this EPL comparison block.
qobs_pot_epl = jnp.full_like(mass_amp_epl, 0.8)


def derive_base_grid_params(x, y, sigma_lum, qobs_lum, sigma_psf, pix_size,
                            nrad=20, nang=10, interp=True, step_input=0.0):
    x = np.asarray(x)
    y = np.asarray(y)
    sigma_lum = np.asarray(sigma_lum)
    qobs_lum = np.asarray(qobs_lum)
    sigma_psf = np.asarray(sigma_psf)
    w = sigma_lum < np.max(np.abs(x))
    qmed = np.median(qobs_lum) if w.sum() < 3 else np.median(qobs_lum[w])
    rell = np.hypot(x, y / qmed)
    psf_convolution = (np.max(sigma_psf) > 0) and (pix_size > 0)
    if (not interp) or ((nrad * nang > x.size) and (not psf_convolution)):
        step = np.min(rell) if step_input == 0 else step_input
        mx = 0.0
    else:
        if psf_convolution:
            step = np.min(sigma_psf) / 4 if step_input == 0 else step_input
            mx = 3 * np.max(sigma_psf) + pix_size / np.sqrt(2)
        else:
            step = np.min(rell) if step_input == 0 else step_input
            mx = 0.0
    rmax = np.max(rell) + mx
    nx = int(np.ceil(rmax / step))
    ny = int(np.ceil(rmax * qmed / step))
    nk = int(np.ceil(mx / step))
    return nx, ny, nk, float(step)


nx_epl, ny_epl, nk_epl, step_epl = derive_base_grid_params(
    x=xbin_epl,
    y=ybin_epl,
    sigma_lum=sigma_lum_epl,
    qobs_lum=qobs_lum_epl,
    sigma_psf=sigmapsf,
    pix_size=pixsize,
    nrad=20,
    nang=10,
    interp=True,
    step_input=0.0,
)

common_kwargs_epl = dict(
    surf_lum=surf_lum_epl,
    sigma_lum=sigma_lum_epl,
    qobs_lum=qobs_lum_epl,
    surf_pot=mass_amp_epl,
    sigma_pot=mass_sig_epl,
    qobs_pot=qobs_pot_epl,
    inc=90.0,
    mbh=0.0,
    distance=distance_epl,
    xbin=xbin_epl,
    ybin=ybin_epl,
    sigmapsf=sigmapsf,
    normpsf=normpsf,
    beta=beta_epl,
    pixsize=pixsize,
    logistic=False,
    moment='zz',
    goodbins=goodbins_epl,
    align='cyl',
    ml=None,
    nrad=20,
    nang=10,
    nlos=1500,
    step=step_epl,
)

scalar_keys = {'inc', 'mbh', 'distance', 'pixsize', 'step'}
pass_keys = {'logistic', 'moment', 'align', 'ml'}


def to_base_kwargs(kwargs):
    base_kwargs = {}
    for k, v in kwargs.items():
        if v is None:
            base_kwargs[k] = v
        elif k in scalar_keys:
            base_kwargs[k] = float(v)
        elif k in pass_keys:
            base_kwargs[k] = v
        elif k == 'goodbins':
            base_kwargs[k] = np.asarray(v, dtype=bool)
        else:
            base_kwargs[k] = np.asarray(v)
    return base_kwargs


base_kwargs_epl = to_base_kwargs(common_kwargs_epl)
jam_base_epl = jam_base(**base_kwargs_epl, plot=False, quiet=True)
sigma_base_epl = np.asarray(jam_base_epl.model)

jam_obj_epl = jam_axi_proj()
jax_kwargs_epl = dict(common_kwargs_epl, nx=nx_epl, ny=ny_epl, nk=nk_epl)
sigma_jax_epl, chi2_epl, flux_epl, ml_epl, vel_epl, vel2_epl = jam_obj_epl.get_kinematics(
    **jax_kwargs_epl,
    quiet=True,
)
sigma_jax_epl = np.asarray(sigma_jax_epl)

eps = np.finfo(float).eps
pct_diff_epl = 100.0 * (sigma_jax_epl - sigma_base_epl) / np.where(np.abs(sigma_base_epl) > eps, np.abs(sigma_base_epl), np.nan)
abs_diff_epl = np.abs(sigma_jax_epl - sigma_base_epl)

x_unique_epl = np.unique(np.asarray(xbin_epl))
y_unique_epl = np.unique(np.asarray(ybin_epl))


def to_grid(values, x_arr, y_arr, x_unique, y_unique):
    values = np.asarray(values)
    grid = np.full((y_unique.size, x_unique.size), np.nan)
    ix = np.searchsorted(x_unique, np.asarray(x_arr))
    iy = np.searchsorted(y_unique, np.asarray(y_arr))
    grid[iy, ix] = values
    return grid


base_grid_epl = to_grid(sigma_base_epl, xbin_epl, ybin_epl, x_unique_epl, y_unique_epl)
jax_grid_epl = to_grid(sigma_jax_epl, xbin_epl, ybin_epl, x_unique_epl, y_unique_epl)
diff_grid_epl = to_grid(pct_diff_epl, xbin_epl, ybin_epl, x_unique_epl, y_unique_epl)

vmin_epl = np.nanmin([np.nanmin(base_grid_epl), np.nanmin(jax_grid_epl)])
vmax_epl = np.nanmax([np.nanmax(base_grid_epl), np.nanmax(jax_grid_epl)])
dmax_epl = np.nanmax(np.abs(diff_grid_epl))

fig, axes = plt.subplots(1, 3, figsize=(12, 4), constrained_layout=False)
fig.subplots_adjust(left=0.004, right=0.996, bottom=0.015, top=0.84, wspace=0.008)

pc0 = axes[0].pcolormesh(x_unique_epl, y_unique_epl, base_grid_epl, shading='auto', cmap='jet', vmin=vmin_epl, vmax=vmax_epl)
axes[0].set_xlabel('x [arcsec]')
axes[0].set_ylabel('y [arcsec]')
cbar0 = fig.colorbar(pc0, ax=axes[0], location='top', fraction=0.03, pad=0.05, format='%.3g')
cbar0.set_label('Base EPL sigma_los [km/s]', fontsize=9, labelpad=8)

pc1 = axes[1].pcolormesh(x_unique_epl, y_unique_epl, jax_grid_epl, shading='auto', cmap='jet', vmin=vmin_epl, vmax=vmax_epl)
axes[1].set_xlabel('x [arcsec]')
axes[1].set_ylabel('')
axes[1].tick_params(labelleft=False)
cbar1 = fig.colorbar(pc1, ax=axes[1], location='top', fraction=0.03, pad=0.05, format='%.3g')
cbar1.set_label('JAX EPL sigma_los [km/s]', fontsize=9, labelpad=8)

norm_epl = TwoSlopeNorm(vmin=-dmax_epl, vcenter=0.0, vmax=dmax_epl) if dmax_epl > 0 else None
pc2 = axes[2].pcolormesh(x_unique_epl, y_unique_epl, np.nan_to_num(diff_grid_epl, nan=0.0), shading='auto', cmap='coolwarm', norm=norm_epl)
axes[2].set_xlabel('x [arcsec]')
axes[2].set_ylabel('')
axes[2].tick_params(labelleft=False)
cbar2 = fig.colorbar(pc2, ax=axes[2], location='top', fraction=0.03, pad=0.05, format='%.3g')
cbar2.set_label('EPL sigma_los difference (%) (JAX - Base)', fontsize=9, labelpad=8)

for ax in axes:
    ax.set_aspect('equal', adjustable='box')
    ax.set_box_aspect(1)
    ax.tick_params(axis='both', which='both', labelsize=8, pad=1)

out_png_epl = 'epl_sigma_los_thetaE1_gamma2_q0.8_zl0.2_zs0.4.png'
fig.savefig(out_png_epl, dpi=240, bbox_inches='tight', pad_inches=0.01)
plt.close(fig)

finite_pct_epl = pct_diff_epl[np.isfinite(pct_diff_epl)]
print(f'saved figure: {out_png_epl}')
print('rotation assumed zero -> sigma_los = Vrms for this block')
print(f'EPL params: thetaE={kwargs_epl_sigma["thetaE"]}, gamma={kwargs_epl_sigma["gamma"]}, q=0.8, zl={kwargs_epl_sigma["zl"]}, zs={kwargs_epl_sigma["zs"]}')
print(f'|Δsigma_los| stats [km/s]: mean={abs_diff_epl.mean():.6g}, median={np.median(abs_diff_epl):.6g}, max={abs_diff_epl.max():.6g}')
if finite_pct_epl.size:
    print(f'Δ% stats: mean={finite_pct_epl.mean():.6g}%, median={np.median(finite_pct_epl):.6g}%, max|%|={np.max(np.abs(finite_pct_epl)):.6g}%')
else:
    print('Δ% stats: all invalid (base model contains zeros everywhere).')
```

```python
import jax
import jax.numpy as jnp
import numpyro
import numpyro.distributions as dist
from numpyro.infer import NUTS, MCMC, init_to_median

key = jax.random.PRNGKey(42)
err_sigma = 20.0  # km/s per pixel

model_true = model
noise = err_sigma * jax.random.normal(key, shape=model_true.shape)
sigma_obs_full = model_true + noise
sigma_obs = sigma_obs_full
x_obs = xbin
y_obs = ybin

jam_eval = partial(jam_obj.get_kinematics,         
                   surf_lum = surf_lum_jackpot,
                    sigma_lum = sigma_lum_jackpot,
                    qobs_lum = qobs_lum_jackpot,
                    inc = 90.0,
                    mbh = 0.0,
                    distance = distance,
                    xbin = xbin,
                    ybin = ybin,
                    sigmapsf = sigmapsf,
                    normpsf = normpsf,
                    pixsize = pixsize,
                    logistic = False,
                    step = 0.05,
                    moment = 'zz',
                    align = 'cyl',
                    ml = None,
                    quiet = True)

def lens_kinematics_model(sigma_obs):

    # --- Priors ---
    thetaE = numpyro.sample("thetaE", dist.Uniform(1.9, 2.1))      # arcsec-ish
    gamma  = numpyro.sample("gamma",  dist.Uniform(1.6, 2.4))      # EPL slope
    beta   = numpyro.sample("beta",   dist.Uniform(-0.5, 0.8))     # JAM anisotropy
    q      = numpyro.sample("q",      dist.Uniform(0.3, 1))         # EPL axis ratio

    beta_vec = jnp.full_like(surf_lum_jackpot, beta)

    mass_amp, mass_sig = mge_epl.decompose(
        thetaE=thetaE, gamma=gamma,
        zl=kwargs_epl["zl"], zs=kwargs_epl["zs"], cosmology=kwargs_epl["cosmology"]
    )
    qobs_pot_local = jnp.full_like(mass_amp, q)

    pred_model, *_ = jam_eval(
                surf_pot = mass_amp,
                sigma_pot = mass_sig,
                beta = beta_vec,
                qobs_pot = qobs_pot_local
    )

    sigma = err_sigma * jnp.ones_like(pred_model)

    with numpyro.plate("pixels", pred_model.size):
        numpyro.sample("obs", dist.Normal(pred_model, sigma), obs=sigma_obs)

nuts = NUTS(lens_kinematics_model, init_strategy=init_to_median(), target_accept_prob=0.8)

backend_now = jax.default_backend()
if GPU:
    # GPU setting requested: 100 warmup / 100 samples / 8 chains.
    # Note: NumPyro valid chain_method is 'vectorized' (not 'cetorized').
    mcmc = MCMC(
        nuts,
        num_warmup=100,
        num_samples=100,
        num_chains=8,
        progress_bar=True,
        chain_method='vectorized',
    )
    print("[MCMC] GPU backend detected (cuda): warmup=100, samples=100, chains=8, chain_method='vectorized'")
else:
    # Lightweight CPU fallback for quick checks.
    mcmc = MCMC(
        nuts,
        num_warmup=10,
        num_samples=10,
        num_chains=1,
        progress_bar=True,
        chain_method='sequential',
    )
    print(f"[MCMC] non-GPU backend ({backend_now}): warmup=10, samples=10, chains=1, chain_method='sequential'")

rng = jax.random.PRNGKey(0)
mcmc.run(rng, sigma_obs=sigma_obs)
```
